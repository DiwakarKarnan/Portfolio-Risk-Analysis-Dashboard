Extracting 15 stocks + 1 index fund last from year 2022. reasen we added index fund because we need something to compare so that we can understand where we are in the market

```python
import yfinance as yf
import pandas as pd
from datetime import datetime
from google.colab import drive

# Mount Google Drive to store processed datasets.
# Power BI reads these files directly from Google Drive,
# enabling the dashboard to display updated data after
# running the notebook and refreshing the report.

# Step 0: Mount Google Drive to store processed data permanently.
drive.mount('/content/drive')

# Step 1: my UK stock tickers + index so python will understand what stock I want
tickers = [
    "RR.L", "BA.L", "IMI.L",
    "HSBA.L", "BARC.L", "LLOY.L",
    "AZN.L", "SGE.L", "REL.L",
    "BP.L", "SHEL.L", "NG.L",
    "ULVR.L", "TSCO.L", "DGE.L",
    "^FTSE"
]

# Step 2: Date range (3+ years) I want everyday daily stock data from 2022 january 1
start_date = "2022-01-01"
end_date = datetime.today().strftime('%Y-%m-%d')

# Step 3: Downloading data from yahoo finance
data = yf.download(tickers, start=start_date, end=end_date)

# Step 4: Extracting close price from each stock
prices = data["Close"]

# Step 5: Clean data (to remove all the missing value)
prices = prices.dropna()

# Step 6: Save the processed data to Google Drive.
# Power BI reads this file when the dashboard is refreshed.
file_path = "/content/drive/MyDrive/YourFolder/uk_portfolio_prices.csv"
prices.to_csv(file_path)

# Step 7: Checking the output
print(prices.head())
print(f"\nData saved successfully at: {file_path} ")
```

Finding the daily return

```python
# Step 8: Calculating daily percentage returns for each asset
returns = prices.pct_change()

# Step 9: first row becomes NaN (normal) sor we removing them
returns = returns.dropna()

# Step 10: Save returns to Google Drive this file important for analysis in power bi
returns_file_path = "/content/drive/MyDrive/YourFolder/uk_portfolio_returns.csv"
returns.to_csv(returns_file_path)

# Step 11: Checking the output
print(returns.head())
print(f"\nReturns saved successfully at: {returns_file_path} ")
```

```python
# Step 12: Creating equal weights for all stocks except FTSE (I have 15(100% investment) each stocks I will invest 6.67% in each stock )
# and FTSE index we are just using to compare the market.
stock_cols = returns.columns.drop("^FTSE")

num_stocks = len(stock_cols)
weights = [1/num_stocks] * num_stocks

# Step 13: Calculating portfolio return (combining all stocks return using weights)
portfolio_returns = returns[stock_cols].dot(weights)

# Step 14: Adding portfolio column in dataframe so I can see individual column and portfolio
returns["Portfolio"] = portfolio_returns

# Step 15: Save updated data
portfolio_file_path = "/content/drive/MyDrive/YourFolder/uk_portfolio_with_portfolio_returns.csv"
returns.to_csv(portfolio_file_path)

# Step 16: Check output
print(returns[["Portfolio"]].head())
print(f"\nPortfolio returns saved at: {portfolio_file_path} ")
```

```python
import numpy as np

# Step 17: Portfolio volatility (calculation volatility how much return flectuated)
portfolio_volatility = returns["Portfolio"].std()

# Step 18: Annualized volatility (converting daily risk to early risk )
annual_volatility = portfolio_volatility * np.sqrt(252)

# Step 19: Portfolio average return (calculating avarage daily return)
portfolio_mean = returns["Portfolio"].mean()

# Step 20: Annualized return (calculating daily return to annual return)
annual_return = portfolio_mean * 252

# Step 21: Sharpe Ratio (return / risk ) how efficient our portfolio is (assuming risk-free rate = 0 for simplicity)
sharpe_ratio = annual_return / annual_volatility

# Step 22: Beta calculation (vs FTSE) comparing our portfolio with market
#Beta tells us how much the portfolio moves compared to the market (FTSE).
#If Beta= 1, portfolio moves similar to the market.
#If Beta is above 1, it is more sensitive than the market.
#If Beta is below 1, it is less sensitive than the market.

cov_matrix = returns[["Portfolio", "^FTSE"]].cov()
beta = cov_matrix.loc["Portfolio", "^FTSE"] / cov_matrix.loc["^FTSE", "^FTSE"]

# Step 23: Printing Return, Volatility, Sharpe Ratio, Beta so I can analyse the data
print("Annual Return:", annual_return)
print("Annual Volatility:", annual_volatility)
print("Sharpe Ratio:", sharpe_ratio)
print("Beta:", beta)
```

```python
# Step 24: Value at Risk (95%) VaR (Worst expected loss (95% confidence))
var_95 = returns["Portfolio"].quantile(0.05) # Finding the worst 5% of portfolio returns and using to calculate Value at Risk (VaR)

print("Value at Risk (95%):", var_95)
```

```python
# Step 25: Rolling volatility (30-day) (risk calculated for over 30 days so first 30 days will be nall (how risk changes overtime))
rolling_vol = returns["Portfolio"].rolling(window=30).std()

# Save it
rolling_vol.to_csv("/content/drive/MyDrive/YourFolder/rolling_volatility.csv")

print(rolling_vol.head())
```

```python
from arch import arch_model

# Step 26: Fit GARCH(1,1) model
garch = arch_model(returns["Portfolio"] * 100, vol='Garch', p=1, q=1)
garch_fit = garch.fit(disp='off')

# Step 27: Get conditional volatility
garch_vol = garch_fit.conditional_volatility

# Step 28: Save it
garch_vol.to_csv("/content/drive/MyDrive/YourFolder/garch_volatility.csv")

# Step 29: Check output
print(garch_vol.head())
print("\nGARCH model completed ")
```

Calculating correlation matrix

```python
# Step 30: Calculating correlation matrix
correlation_matrix = returns.drop(columns=["Portfolio"]).corr()

correlation_file_path = "/content/drive/MyDrive/YourFolder/correlation_matrix.csv"
correlation_matrix.to_csv(correlation_file_path)

print(correlation_matrix)
print(f"Correlation matrix saved at: {correlation_file_path}")
```

```python
metrics = pd.DataFrame({
    "Metric": ["Annual Return", "Annual Volatility", "Sharpe Ratio", "Beta", "VaR 95%"],
    "Python Result": [annual_return, annual_volatility, sharpe_ratio, beta, var_95]
})

metrics.to_csv("/content/drive/MyDrive/YourFolder/python_metrics_summary.csv", index=False)
```
